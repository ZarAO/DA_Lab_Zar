## Training a Linear Regression Model

```python
X = LabData[['x']]
y = LabData['y']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
ls = float(metrics.mean_squared_error(y_test, predictions))
metrics.mean_squared_error(y_test, predictions)
```

```python
class Linear():
    
    def __init__(self, iterCount=10000):
        self.a = 0
        self.b = 0
        self.count = iterCount
    
    def fit(self, X, y, ls):
        lenghRange = float(len(X))

        for i in range(self.count): 
            Y_approx = self.a*X + self.b
            self.a -= ls * (-2/lenghRange) * sum(X * (y - Y_approx))
            self.b -= ls * (-2/lenghRange) * sum(y - Y_approx)
    
    def Get(self):
        return self.a, self.b
```

```python
lr = Linear()
X = LabData['x']
lr.fit(X,y,ls)
lr.Get()
```
